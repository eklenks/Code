```python
import quandl
import matplotlib.pyplot as plt

quandl.ApiConfig.api_key = 'YOUR_API_KEY'

overnight_rate = quandl.get('BOC/V80691334')
one_year_bond_yield = quandl.get('BOC/V80691321')

plt.figure(figsize=(14, 7))
plt.plot(overnight_rate.index, overnight_rate['Value'], label='Overnight Rate', color='blue')
plt.plot(one_year_bond_yield.index, one_year_bond_yield['Value'], label='1-Year Bond Yield', color='green')

plt.title('Canadian Overnight Rate and 1-Year Bond Yield')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.legend()
plt.grid(True)

plt.show()
```
