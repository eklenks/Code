$$SWF \equiv E[u(y,l,G)] = \int_{0}^{\infty} f(w)u(y(w),l(w),G) \, dw$$


```python
import numpy as np
from scipy.stats import lognorm
from scipy.optimize import minimize_scalar
# Define the utility function
def utility(y, l, G, eta, gamma):
    return y**eta * l**gamma * G**(1-eta-gamma)
# Define the budget constraint
def y_wage(w, h):
    return w * h
# Define the log-normal distribution function
def lognormal_pdf(w, mu, sigma_sq):
    return 1 / (w * np.sqrt(2 * np.pi * sigma_sq)) * np.exp(-((np.log(w) - mu)**2) / (2 * sigma_sq))
    
# Compute SWF for given parameters
def compute_swf(eta, gamma, sigma_sq, mu, h, G):
    # Create an array of wage values
    w_values = np.linspace(0.01, 10, 1000)
    # Calculate utility for each wage value and multiply by the PDF
    integrand = utility(y_wage(w_values, h), h, G, eta, gamma) * lognormal_pdf(w_values, mu, sigma_sq)
    # Integrate using trapezoidal rule
    swf = np.trapz(integrand, w_values)
    return swf
# Define different parameter sets
parameter_sets = [
    {'eta': 0.5, 'gamma': 0.3, 'sigma_sq': 0.5, 'mu': -0.5 / 2, 'h': 0.5, 'G': 1},
    {'eta': 0.5, 'gamma': 0.3, 'sigma_sq': 0.8, 'mu': -0.8 / 2, 'h': 0.5, 'G': 1},
    {'eta': 0.5, 'gamma': 0.3, 'sigma_sq': 0.5, 'mu': -0.5, 'h': 0.5, 'G': 1},
    {'eta': 0.3, 'gamma': 0.2, 'sigma_sq': 0.5, 'mu': -0.5 / 2, 'h': 0.7, 'G': 1}
]
# Print table header
print("{:<5} {:<5} {:<5} {:<8} {:<8} {:<5} {:<8}".format("Case", "eta", "gamma", "var", "mu", "h", "SWF"))
# Compute SWF for each parameter set
for i, params in enumerate(parameter_sets):
    swf = compute_swf(**params)
    print("{:<5} {:<5} {:<5} {:<8.2f} {:<8.2f} {:<5} {:<8.4f}".format(i+1, params['eta'], params['gamma'], params['sigma_sq'], params['mu'], params['h'], swf))
```

```python
# adjust swf to include pi parameter
def compute_swf_pi(eta, gamma, sigma_sq, mu, h, G, pi):
    # Create an array of wage values
    w_values = np.linspace(0.01, 10, 1000)
    # Calculate utility for each wage value and multiply by the PDF
    integrand = utility(y_wage(w_values, h), h, G, eta, gamma) * lognormal_pdf(w_values, mu, sigma_sq)
    # Integrate using trapezoidal rule
    average_utility = np.trapz(integrand, w_values)
    # Adjust for preference type proportion
    average_utility *= pi
    return average_utility
# Compute certainty equivalent income for given parameters and preference type
def compute_certainty_equivalent_income(eta_L, eta_H, gamma, sigma_sq, mu, h, G, pi):
    # Compute average utility for low and high preference types
    avg_utility_L = compute_swf_pi(eta_L, gamma, sigma_sq, mu, h, G, pi)
    avg_utility_H = compute_swf_pi(eta_H, gamma, sigma_sq, mu, h, G, 1-pi)
    # Compute certainty equivalent income for each type
    y_ce_L = (avg_utility_L / (1 - eta_L))**(1 / eta_L)
    y_ce_H = (avg_utility_H / (1 - eta_H))**(1 / eta_H)
    return y_ce_L, y_ce_H
# Define different parameter sets
parameter_sets = [
    {'pi': 0.5, 'gamma': 0.3, 'sigma_sq': 0.5, 'mu': -0.5 / 2, 'h': 0.5, 'G': 1},
    {'pi': 0.5, 'gamma': 0.3, 'sigma_sq': 0.8, 'mu': -0.8 / 2, 'h': 0.5, 'G': 1},
    {'pi': 0.5, 'gamma': 0.3, 'sigma_sq': 0.5, 'mu': -0.5, 'h': 0.5, 'G': 1},
    {'pi': 0.5, 'gamma': 0.2, 'sigma_sq': 0.5, 'mu': -0.5 / 2, 'h': 0.7, 'G': 1}
]
# Print table header
print("Parameter Values\t\t\t\tCertainty Equivalent Income (y_ce)")
print("π\tγ\tσ^2\t\tµ\t\th\t\t\tη_L=0.4\t\tη_H=0.6")
# Compute and print certainty equivalent income for each parameter set
for i, params in enumerate(parameter_sets):
    pi = params['pi']
    gamma = params['gamma']
    sigma_sq = params['sigma_sq']
    mu = params['mu']
    h = params['h']
    G = params['G']
    eta_L = 0.4
    eta_H = 0.6
    y_ce_L, y_ce_H = compute_certainty_equivalent_income(eta_L, eta_H, gamma, sigma_sq, mu, h, G, pi)
    print("{:.1f}\t{:.1f}\t{:.1f}\t\t{:.1f}\t\t{:.1f}\t\t\t{:.4f}\t\t{:.4f}".format(pi, gamma, sigma_sq, mu, h, y_ce_L, y_ce_H))
```

$$ G = \int_{0}^{\infty} f(w) \tau w h \, dw $$
$$ \max_{\tau} SWF(\tau) \equiv \mathbb{E} [u(y(w), l(w), G)] $$



```python
# Define the budget constraint with tax parameter tau
def y_wage(w, h, tau):
    return (1 - tau) * w * h
# Compute G for given parameters and tax tau
def compute_G(tau, h, mu, sigma_sq):
    # Create an array of wage values
    w_values = np.linspace(0.01, 10, 1000)
    # Calculate G for each wage value and multiply by the PDF
    G = np.trapz(lognormal_pdf(w_values, mu, sigma_sq) * tau * w_values * h, w_values)
    return G
# Compute SWF for given parameters and tax tau
def compute_swf(tau, eta, gamma, sigma_sq, mu, h):
    # Compute G
    G = compute_G(tau, h, mu, sigma_sq)
    # Create an array of wage values
    w_values = np.linspace(0.01, 10, 1000)
    # Calculate utility for each wage value and multiply by the PDF
    integrand = utility(y_wage(w_values, h, tau), h, G, eta, gamma) * lognormal_pdf(w_values, mu, sigma_sq)
    # Integrate using trapezoidal rule
    swf = np.trapz(integrand, w_values)
    return swf
# Define the log-normal distribution function
def lognormal_pdf(w, mu, sigma_sq):
    return 1 / (w * np.sqrt(2 * np.pi * sigma_sq)) * np.exp(-((np.log(w) - mu)**2) / (2 * sigma_sq))
# Define different parameter sets
parameter_sets = [
    {'eta': 0.5, 'gamma': 0.3, 'sigma_sq': 0.5, 'mu': -0.5 / 2, 'h': 0.5},
    {'eta': 0.5, 'gamma': 0.3, 'sigma_sq': 0.8, 'mu': -0.8 / 2, 'h': 0.5},
    {'eta': 0.5, 'gamma': 0.3, 'sigma_sq': 0.5, 'mu': -0.5, 'h': 0.5},
    {'eta': 0.3, 'gamma': 0.2, 'sigma_sq': 0.5, 'mu': -0.5 / 2, 'h': 0.7}
]
# Print table header
print("{:<5} {:<8} {:<8} {:<8}".format("Case", "Tau", "G", "SWF"))
# Compute maximum SWF for each parameter set
for i, params in enumerate(parameter_sets):
    # Define a lambda function for negative SWF to use minimize_scalar
    negative_swf = lambda tau: -compute_swf(tau, **params)
    # Find optimal tau that maximizes SWF
    optimal_tau = minimize_scalar(negative_swf, bounds=(0, 1), method='bounded').x
    # Compute G
    G = compute_G(optimal_tau, params['h'], params['mu'], params['sigma_sq'])
    # Compute maximum SWF using optimal tau
    max_swf = -negative_swf(optimal_tau) # negate back to get positive SWF
    print("{:<5} {:<8.4f} {:<8.4f} {:<8.4f}".format(i+1, optimal_tau, G, max_swf))
```
